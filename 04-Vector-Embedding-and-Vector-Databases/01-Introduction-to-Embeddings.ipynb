## What are Embeddings?

Think of embeddings as a way to translate words into a language that computers understand

## Setup and Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Understanding Embeddings

```python
word_embeddings = {
    "cat": [0.8, 0.6],
    "kitten": [0.75, 0.65],
    "dog": [0.7, 0.3],
    "puppy": [0.65, 0.35],
    "car": [-0.5, 0.2],
    "truck": [-0.45, 0.15],
}
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

for word, coordinates in word_embeddings.items():
    ax.scatter(coordinates[0], coordinates[1], s=100)
    ax.annotate(word, (coordinates[0], coordinates[1]), xytext=(5,5),
                textcoords="offset points")
    
ax.set_xlabel("Dimension 1")
ax.set_ylabel("Dimension 2")
ax.set_title("Simplified Word Embeddings")
ax.grid(visible=True, alpha=0.3)

plt.tight_layout()
```

## Measuring Similarity

```python
def cosine_similarity(vector1, vector2):
    """ 
    Cosine Similarity measures the angel between two vectors.
    - Result close to 1: Very similiar
    - Result close to 0: Not related
    - Result close to -1: Opposite meaning
    """

    dot_product = np.dot(vector1, vector2)
    norm_vector1 = np.linalg.norm(vector1)
    norm_vector2 = np.linalg.norm(vector2)

    return dot_product/(norm_vector1 * norm_vector2)
```

```python
cat_vector = [0.8, 0.6, 0.3]
kitten_vector = [0.75, 0.65, 0.35]
car_vector = [-0.5, 0.2, 0.1]
dog_vector = [0.7, 0.3, 0.7]

cat_kitten_similarity = cosine_similarity(vector1=cat_vector, vector2=kitten_vector)
cat_car_similarity = cosine_similarity(vector1=cat_vector, vector2=car_vector)
cat_dog_similarity = cosine_similarity(vector1=cat_vector, vector2=dog_vector)

print(f"Similarity Score of Cat and Kitten is: {cat_kitten_similarity}")
print(f"Similarity Score of Cat and Car is: {cat_car_similarity}")
print(f"Similarity Score of Cat and Dog is: {cat_dog_similarity}")

```
